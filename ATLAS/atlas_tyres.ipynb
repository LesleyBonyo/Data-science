# ATLAS TYRES


The excel sheet to be explored and analysed is from Atlas Tyres. It captures sales of their different models of tyres from January to July 2019. Each work sheet contains sales data for each model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib.dates import DateFormatter

%matplotlib inline
```

```python
atlas = pd.read_excel("SALES BREAKDOWN.xlsx", None) #read excel sheet but open no particular sheet
atlas.keys() #sheets on the excel file represent different models of tyres
```

```python
#atlas = atlas[0:37] #ewmove columns wiht NaN values
sheet1 = atlas['155R12C ATL'] #view the first sheet
sheet1.head()
```

```python
sheet1.shape
```

```python
sheet1.tail()
```

```python
sheet1
```

```python
sheet1 = sheet1[1:37] #include second row to 36th row
sheet1
```

```python
date_sheet1 = sheet1.loc[:,['DATE']]
```

```python
qty_sheet1 = sheet1.loc[:,['QTY']]
```

```python
plt.style.use('ggplot')
plt.plot(date_sheet1, qty_sheet1, color = 'blue') # comparing sales for tyre 155R12C ATL from January to July 2019
plt.title('Comparing sales for Tyre 155R12C ATL from January to July 2019')
plt.xlabel('Date')
plt.ylabel('Quantity of Sales')
```
